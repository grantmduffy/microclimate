```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
# z = np.sin(np.linspace(0, np.pi * 2 * 4, 256)) * 0.5 + 0.5
z = np.linspace(-1, 1, 256) ** 2
w = np.clip(-np.linspace(-2, 2, 256) ** 2 + 1, 0, 1)
# z = np.zeros(256)
# w = np.ones(256)

plt.figure()
cmap = plt.get_cmap('plasma')
for i, a in enumerate(np.linspace(0, 1, 10000)):
    wz = z + w
    if i % 10 == 0:
        plt.plot(wz[1:-1], c=cmap(a))
    flux_in_left = np.clip(wz[:-2] - wz[1:-1], -w[1:-1] / 2, w[:-2] / 2)
    flux_in_right = np.clip(wz[2:] - wz[1:-1], -w[1:-1] / 2, w[2:] / 2)
    flux = flux_in_left + flux_in_right
    w[1:-1] += 1 * flux
    
plt.plot(z[1:-1], 'k')
plt.legend()
plt.show()
```
